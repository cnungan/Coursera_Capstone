### 1) Getting the Table from Wikipedia

Here, I will use Beautiful Soup to learn different libraries and to spice things up!
Let's start by importing necessary libraries.

```python
import pandas as pd
import requests
import numpy as np
```

```python
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
!conda install -c anaconda beautifulsoup4 --yes

from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(website_url,'html5lib')
print(soup.prettify())
```

```python
#finding our table from the html
table = soup.find('table',{'class':'wikitable sortable'})
print(table)
```

```python
#let's define lists of our attributes 
Postal_Code = []
Borough = []
Neighborhood = []

#and loop through data to extract rows and append them to the corresponding lists
for row in table.find_all('tr')[1:]:
    Postal_Code_D = row.find_all('td')[0].text  
    Postal_Code.append(Postal_Code_D)
print(Postal_Code)
for row in table.find_all('tr')[1:]:
    Borough_cell = row.find_all('td')[1]
    Borough.append(Borough_cell.text)
print(Borough)
for row in table.find_all('tr')[1:]:
    Neighborhood_cell = row.find_all('td')[2]
    Neighborhood.append(Neighborhood_cell.text)
print(Neighborhood)
```

```python
#eliminate the \n
Postal_Code = list(map(lambda x:x.strip(),Postal_Code))
Borough = list(map(lambda x:x.strip(),Borough))
Neighborhood = list(map(lambda x:x.strip(),Neighborhood))
```

```python
#and populate the df1 dataframe with the lists 
df1 = pd.DataFrame()
df1['Postal_Code'] = Postal_Code
df1['Borough'] = Borough
df1['Neighborhood'] = Neighborhood
df1.head()
```

```python
#to drop the 'Not assigned' values, first I extracted the indexnames
indexnames = df1[df1['Borough'] == 'Not assigned'].index

#then drop the rows with 'Not assigned' values with the help of their indexnames
df1.drop(indexnames , inplace=True)
df1.head()
```

```python
#reset the index of the cleaned dataframe
df1 = df1.reset_index(drop = True)
df1
```

```python
#shape of our dataframe
df1.shape
```

```python
LIMIT = 100
RADUIUS = 500

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

  ##### Finding 5 most common venues for each neighborhood

```python
num_top_venues = 5

for hood in grouped_toronto['Neighborhood']:
    print("----"+hood+"----")
    temp = grouped_toronto[grouped_toronto['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
#descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#top 10 venues for each neighborhood

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted2 = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted2['Neighborhood'] = grouped_toronto['Neighborhood']

for ind in np.arange(grouped_toronto.shape[0]):
    neighborhoods_venues_sorted2.iloc[ind, 1:] = return_most_common_venues(grouped_toronto.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted2.head()
```

#### 4) Neighborhood Clustering

```python
# set number of clusters
kclusters = 8

grouped_toronto_clustering = grouped_toronto.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(grouped_toronto_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
# add clustering labels
neighborhoods_venues_sorted2.insert(0, 'Cluster Labels', kmeans.labels_)

merged_toronto = df_toronto

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
merged_toronto = merged_toronto.join(neighborhoods_venues_sorted2.set_index('Neighborhood'), on='Neighborhood')

merged_toronto.head() # check the last columns!
```

Finally, let's visualize the resulting clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(merged_toronto['Latitude'], merged_toronto['Longitude'], merged_toronto['Neighborhood'], merged_toronto['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
